## l2 norm and AMP

```python
rand_fp16 = tf.random.uniform(
    (1,1000000),
    minval=-1,
    maxval=1,
    dtype=tf.dtypes.float16,
    seed=None,
)

print("L2-normalize in FP16 Then Convert to FP32")
rand_fp16_norm = tf.math.l2_normalize(rand_fp16)
print(f"Norm: {tf.norm(rand_fp16_norm).numpy()}")
rand_fp16_norm_fp32 = tf.cast(rand_fp16_norm, tf.dtypes.float32)
print(f"Norm: {tf.norm(rand_fp16_norm_fp32).numpy()}")

print("\nFirst Convert to FP32 then L2-normalize")
rand_fp16_fp32 = tf.cast(rand_fp16, tf.dtypes.float32)
rand_fp16_fp32_norm = tf.math.l2_normalize(rand_fp16_fp32)
print(f"Norm: {tf.norm(rand_fp16_fp32_norm).numpy()}")

print(f"\nMax Absolute Difference: {tf.reduce_max(tf.math.abs(rand_fp16_norm_fp32-rand_fp16_fp32_norm)).numpy()}")
```

```python
rand_fp32 = tf.random.uniform(
    (100,),
    minval=-1,
    maxval=1,
    dtype=tf.dtypes.float32,
    seed=None,
    name=None
)
rand_fp32_norm = tf.math.l2_normalize(rand_fp32)

rand_fp16 = tf.cast(rand_fp32, tf.dtypes.float16)
rand_fp16_norm = tf.math.l2_normalize(rand_fp16)
rand_fp16_norm_fp32 = tf.cast(rand_fp16_norm, tf.dtypes.float32)

print(tf.reduce_max(tf.math.abs(rand_fp32_norm-rand_fp16_norm_fp32)).numpy())
```
